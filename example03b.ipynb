## Inladen van externe libraries

```python
from example03 import *
import matplotlib.pyplot as plt
```

## Instellen van sensor object en simulatie

```python
fixed_pulses = 96

encoder_data = EncoderData()
fixed_pulses_sensor = FixedPulsesSensor(encoder_data, fixed_pulses=fixed_pulses)

simulatie_fs = 250
simulatie = Simulatie(
    signaal_observers=[fixed_pulses_sensor],
    lengte=30.,
    dt=float(1/simulatie_fs)
)

resultaat = simulatie.simuleer()
```

## plotten van simulatie

```python
t = resultaat.tijdstippen()
dt_dpulses = resultaat.signalen()

plt.figure(figsize=(14,6))
plt.scatter(t, dt_dpulses)
plt.grid()
plt.title(f'Delta tijd per {fixed_pulses} pulses')
plt.xlabel("Tijd [s]")
plt.xlim((0, 30))
plt.ylabel("Delta tijd [s]")
```
