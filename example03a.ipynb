```python
from example03 import *
import matplotlib.pyplot as plt
```

```python
fs = 50

encoder_data = EncoderData()
fixed_tijd_sensor = FixedTijdSensor(encoder_data, delta_tijd=1/fs)

simulatie = Simulatie(
    signaal_observers=[fixed_tijd_sensor],
    lengte=30.,
    dt=float(1/fs)
)

resultaat = simulatie.simuleer()
```

```python
t = resultaat.tijdstippen()
nr_pulses = resultaat.signalen()

plt.figure(figsize=(14,6))
plt.scatter(t, nr_pulses)
plt.grid()
plt.title("Aantal getelde pulses over tijd")
plt.xlabel("Tijd [s]")
plt.xlim((0, 30))
plt.ylabel("Pulses [-]")
```
