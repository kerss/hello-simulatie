## Inladen van externe libraries

```python
from example03 import *
import matplotlib.pyplot as plt
```

## Instellen van sensor object en simulatie

```python
sensor_delta_tijd = 0.100

encoder_data = EncoderData()
fixed_tijd_sensor = FixedTijdSensor(encoder_data, delta_tijd=sensor_delta_tijd)

simulatie = Simulatie(
    signaal_observers=[fixed_tijd_sensor],
    lengte=30.,
    dt=float(sensor_delta_tijd)
)

resultaat = simulatie.simuleer()

t = resultaat.tijdstippen()
nr_pulses = resultaat.signalen()
```

## Plotten van simulatie

```python
plt.figure(figsize=(14,6))
plt.scatter(t, nr_pulses)
plt.grid()
plt.title("Aantal getelde pulses over tijd")
plt.xlabel("Tijd [s]")
plt.xlim((0, 30))
plt.ylabel("Pulses [-]")
```
